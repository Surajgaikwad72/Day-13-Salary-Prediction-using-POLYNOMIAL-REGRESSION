## Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Loading dataset

```python
df = pd.read_csv("dataset Salary.csv")
```

```python
df
```

```python
df.shape
```

## Segregate the dataset into X and Y 

```python
X = df.iloc[:,:-1].values
X
```

```python
Y = df.iloc[:,-1].values
Y
```

## Training the model with linear regression

```python
from sklearn.linear_model import LinearRegression
modelLR = LinearRegression()
modelLR.fit(X,Y)
```

## Visualize the dataset

```python
plt.scatter(X,Y,color="red")
plt.plot(X,modelLR.predict(X))
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Linear Regression")
plt.show()
```

# Step 6 as we can see the Linear Regression is not performing well because not all data points are covered 

### Converting the X into ploy x

```python
from sklearn.preprocessing import PolynomialFeatures
modelPR = PolynomialFeatures(degree = 4 )
xPloy = modelPR.fit_transform(X)
```

```python
modelPLR = LinearRegression()
modelPLR.fit(xPloy,Y)
```

## Visualize the dataset by polynomial Regression

```python
plt.scatter(X,Y,color="red")
plt.plot(X,modelPLR.predict(modelPR.fit_transform(X)))
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Polynomial Regression")
plt.show()
```

## Predicting

```python
a = 8
PredictModel = modelPLR.predict(modelPR.fit_transform([[a]]))
print("Salary of the person with the level {0} is {1}".format(a,PredictModel))
```
